```python
import pandas as pd
```

```python
# Reading the csv file
df_population_raw = pd.read_csv(r"C:\Users\krishna rawal\OneDrive\Desktop\population_total.csv")
```

```python
# dropping null values
df_population_raw.dropna()
```

```python
# dropping null values
df_population_raw.dropna(inplace=True)
```

```python
# MAKING A pivot tabledf
df_pivot = df_population_raw.pivot(index='year' , columns='country' , values='population' )
```

```python
# selecting some countries
df_pivot = df_pivot[['United States' , 'India' , 'China' , 'Indonesia' , 'Brazil']]
```

```python
# showing pivot table
df_pivot
```

## 2 Lineplot

```python
df_pivot.plot(kind='line' )
```

```python
df_pivot.plot(kind='line', xlabel='Year' , ylabel='Population' , title='population (1955-2020)' )
```

```python
df_pivot.plot(kind='line', xlabel='Year' , ylabel='Population' ,\
              title='population (1955-2020)',
             figsize=(8,4) )
```
